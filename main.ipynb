```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def grayimg():
    root = os.getcwd()
    imgpath = os.path.join(root, 'bunga.webp')
    img = cv.imread(imgpath, cv.IMREAD_GRAYSCALE)

    # Menampilkan gambar grayscale 
    plt.figure()
    plt.imshow(img, cmap='gray')
    plt.title('Gambar Grayscale')
    plt.axis('off')

    # Menghitung histogram
    hist = cv.calcHist([img], [0], None, [256], [0, 256])

    # Menampilkan histogram
    plt.figure()
    plt.plot(hist, color='black')
    plt.title('Histogram Gambar Grayscale')
    plt.xlabel('Piksel Intensitas')
    plt.xlim([0, 256])
    plt.ylabel('Jumlah Piksel')
    plt.grid(True)
    plt.show()

    # Menghitung statistik histogram
    mean = np.mean(img)
    print(f"Rata-rata: {mean:.2f}")
    stddev = np.std(img)
    print(f"Standar Deviasi: {stddev:.2f}")
    var = np.var(img)
    print(f"Variansi: {var:.2f}")

if __name__ == '__main__':
    grayimg()
```

Image Histogram RGB

```python
def colorimg():
    root = os.getcwd()
    imgpath = os.path.join(root, 'alam.jpg')
    img = cv.imread(imgpath)
    rgb_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)


    # Menampilkan gambar warna
    plt.figure()
    plt.imshow(rgb_img)
    plt.title('Gambar Warna')
    plt.axis('off')
    plt.show()

    # Histogram untuk setiap channel warna
    colors = ['red', 'green', 'blue']
    plt.figure()
    for i in range(len(colors)):
        hist = cv.calcHist([rgb_img], [i], None, [256], [0, 256])
        plt.plot(hist, color = colors[i])
    
    plt.title('Histogram Gambar Warna')
    plt.xlim([0, 256])
    plt.xlabel('Piksel Intensitas')
    plt.ylabel('Jumlah Piksel')
    plt.grid(True)
    plt.show()

    # Menghitung statistik histogram untuk setiap channel warna
    for i, color in enumerate(colors):
        mean = np.mean(rgb_img[:, :, i])
        print(f"Rata-rata {color}: {mean:.2f}")
        stddev = np.std(rgb_img[:, :, i])
        print(f"Standar Deviasi {color}: {stddev:.2f}")
        var = np.var(rgb_img[:, :, i])
        print(f"Variansi {color}: {var:.2f}")

if __name__ == '__main__':
    colorimg()
```
